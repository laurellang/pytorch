(1)nn.Module是什么？

它是一种神经网络基本骨架，以下面为例：

```python
#定义卷积神经网络
class Tudui(nn.Module):
    def __init__(self):
        super(Tudui, self).__init__()
        self.conv1 = Conv2d(in_channel = 3, out_channel = 6, kernel_size = 3)

    def forward(self, x):
        x = self.conv1(x)
        return x
```

1、nn.Module放在括号内，表示Tudui这个类继承了Module

2、super是什么意思？
super（）是python中调用父类的方法。super(Tudui, self).__init__()表示调用Tudui的父类（nn.Module）的__init__()方法。

下面有一个例子来说明：

```python
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal __init__ called, name={name}")
    
    def speak(self):
        print("动物叫")

class Dog(Animal):
    def __init__(self, name, breed):
        # 如果不调用super，父类的__init__不会执行
        # self.name 不会被设置！
        super().__init__(name)  # 调用Animal.__init__(name)
        self.breed = breed
        print(f"Dog __init__ called, breed={breed}")

dog = Dog("旺财", "金毛")
# 输出：
# Animal __init__ called, name=旺财
# Dog __init__ called, breed=金毛

print(dog.name)   # 旺财 （来自父类）
print(dog.breed)  # 金毛 （来自子类）
```

（续）2、那么super(Tudui, self).__init__()初始化了什么呢？

```python
class Tudui(nn.Module):
    def __init__(self):
        # 如果不写这行，nn.Module的初始化代码不会执行
        super(Tudui, self).__init__()  # 必须调用！
        
        # nn.Module.__init__()做了这些重要事情：
        # 1. 初始化_parameters字典（存权重参数）
        # 2. 初始化_modules字典（存子模块）
        # 3. 初始化其他内部状态
        
        # 然后你才能添加自己的层
        self.conv1 = nn.Conv2d(3, 6, 3)
        # PyTorch会自动把conv1加入到_modules字典中
```
