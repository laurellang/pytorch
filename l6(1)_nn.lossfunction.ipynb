L1Loss():

class torch.nn.L1Loss(size_average=None, reduce=None, reduction='mean')[source]


MSELoss（）：

class torch.nn.MSELoss(size_average=None, reduce=None, reduction='mean')

```python
import torch
from torch.nn import L1Loss, MSELoss

inputs = torch.tensor([1, 2, 3], dtype=torch.float32)
targets = torch.tensor([1, 2, 5], dtype=torch.float32)

inputs = torch.reshape(inputs, (1, 1, 1,3))
targets = torch.reshape(targets, (1, 1, 1, 3))

loss1 = L1Loss(reduction="sum") #要先创建实例
loss2 = MSELoss()
result1 = loss1(inputs, targets)
result2 = loss2(inputs, targets)
print(result1)
print(result2)
```

<span style="font-size:40px">交叉熵</span>

CrossEntropyLoss():

class torch.nn.CrossEntropyLoss(weight=None, size_average=None, ignore_index=-100, reduce=None, reduction='mean', label_smoothing=0.0)

计算公式如下：


$$
\mathcal{L}(x, \text{class}) = -\ln \left( \frac{\exp(x[\text{class}])}{\sum\limits_{j} \exp(x[j])} \right) = -x[\text{class}] + \ln \left( \sum\limits_{j} \exp(x[j]) \right)
$$


<span style="color:lightblue">**对“交叉”二字的理解**：</span>

想象在完成一场考试后，需要用标准答案来批改自己的答案，故这是两份答案“交叉”的情况。

那么在计算交叉熵的时候，需要用真实分布的权重，乘上预测分布的信息量（其实只是log），这样就完成了“批改答案”这个过程

标准交叉熵定义如下：


$$H(P,Q) = -\sum_i P(i)\log Q(i)$$

【Q(i)s是经过softmax函数处理后，用分数转化的概率分布，取得是以e为底的对数】

<span style="color:lightgreen">上述两个公式是等价的，因为第二个公式只有当i=class的时候p(i)才会为1，也就是说只有Q(i)，而Q(i)就是</span>
$$ \frac{\exp(x[\text{class}])}{\sum\limits_{j} \exp(x[j])}$$

```python
x = torch.tensor([0.1, 0.2, 0.3])
y = torch.tensor([1])
x = torch.reshape(x, (1, 3))

from torch import nn
loss_cross = nn.CrossEntropyLoss()
result_cross = loss_cross(x, y)
'''
loss_cross的参数应该填什么？
input：（C）——单样本、（N,C)——批量样本、（N, C, d1, d2, ...)——批量多维样本
target：同上
'''

print(result_cross)
```
